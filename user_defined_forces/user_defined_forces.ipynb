```python
import swiftest
```

```python
sim_orig = swiftest.Simulation(simdir="no_extra", verbose=False)
sim_orig.add_solar_system_body(["Sun", "Jupiter", "Saturn", "Uranus", "Neptune"])
sim_orig.run(tstop=1e5, dt=2.0, tstep_out=100, dump_cadence=0, extra_force=False)
```

```python
sim_orig.data.inc.where(sim_orig.data.particle_type != "Central Body", drop=True).plot(x="time", hue="name")
```

```python
sim_jup = swiftest.Simulation(simdir="with_extra",verbose=False)
sim_jup.add_solar_system_body(["Sun", "Jupiter", "Saturn", "Uranus", "Neptune"])
sim_jup.run(tstop=1e5, dt=2.0, tstep_out=100, dump_cadence=0, extra_force=True)
```

```python
sim_jup.data.inc.where(sim_jup.data.particle_type != "Central Body", drop=True).plot(x="time", hue="name")
```

```python
(sim_jup.data["inc"].sel(name="Jupiter") - sim_orig.data.inc.sel(name="Jupiter")).plot(x="time")
```
